# TASD-Dataset: Text-based Early ASD Detection

#### About Dataset
The TASD-Dataset is tailored for early ASD detection in toddlers, providing detailed textual sequences that describe the daily situations of toddlers both with and without ASD. It incorporates essential ASD assessment features such as Attention Response, Word Repetition, Emotional Empathy, and introduces new features including Noise Sensitivity, Sharing Interest, Sign Communication, and Tiptoe Flapping. Each feature is intricately linked with specific toddler behaviors, offering nuanced insights for parents and aiding in the identification of ASD-related signals. By emphasizing these features, the dataset facilitates the development of machine learning models to explore behavioral markers crucial for early ASD risk detection.

### Top Most Common Stopwords

```python
stop = set(stopwords.words('english'))

# Get all the word tokens in dataframe for Disaster and Non-Disaster
corpus0 = [] # Non-Disaster
[corpus0.append(word.lower()) for tweet in data[data.ASD == "Non-ASD"].Text for word in word_tokenize(tweet)]
corpus1 = [] # Disaster
[corpus1.append(word.lower()) for tweet in data[data.ASD == "ASD"].Text for word in word_tokenize(tweet)]
```

```python
def count_top_stopwords(corpus):
    stopwords_freq = {}
    for word in corpus:
        if word in stop: 
            if word in stopwords_freq:
                stopwords_freq[word] += 1
            else:
                stopwords_freq[word] = 1
    topwords = sorted(stopwords_freq.items(), key=lambda item: item[1], reverse=True)[:10] # get the top 10 stopwords
    x,y = zip(*topwords) # get key and values
    return x,y
```

```python
x0,y0 = count_top_stopwords(corpus0)
x1,y1 = count_top_stopwords(corpus1)

plt.figure(figsize=(15,4))
plt.subplot(1,2,1)
plt.bar(x0,y0, color=color[0])
plt.title('Top Stopwords for Non-ASD')
plt.subplot(1,2,2)
plt.bar(x1,y1, color=color[1])
plt.title('Top Stopwords for  ASD')
```

### Meta-Feature Engineering

Here, we extract some features from the data that might give us some idea about whether it is a disaster or not. The purpose of this is to build a feature-based model and use it as part of an ensemble model to improve the predictions of the sequence model. Although it might not perform well on its own, it can  actually boost the performance when combined with other models.

* polarity - range of \[-1,1] where 1 denotes positivity and -1 denotes negativity
* subjectivity - range of \[0,1] where 1 denotes personal opinions and 0 denotes factual info
* exclaimation_num - number of exclamation marks in tweet
* questionmark_num - number of question marks in tweet
* url_num - number of urls in tweet
* hash_num - number of hashtags (#) in tweet
* mention_num - number of mentions (@) in tweet
* contraction_num - number of contractions (e.g I'm, we're, we've)

```python
# polarity and subjectivity
data['polarity'] = [TextBlob(tweet).sentiment.polarity for tweet in data.Text]
data['subjectivity'] = [TextBlob(tweet).sentiment.subjectivity for tweet in data.Text]

# exclaimation and question marks
data['exclaimation_num'] = [tweet.count('!') for tweet in data.Text]
data['questionmark_num'] = [tweet.count('?') for tweet in data.Text]
```

```python
# count number of hashtags and mentions
def count_url_hashtag_mention(text):
    urls_num = len(re.findall('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', text))
    word_tokens = text.split()
    hash_num = len([word for word in word_tokens if word[0] == '#' and word.count('#') == 1]) # only appears once in front of word 
    mention_num = len([word for word in word_tokens if word[0] == '@' and word.count('@') == 1]) # only appears once in front of word 
    return urls_num, hash_num, mention_num
```

```python
url_num, hash_num, mention_num = zip(*[count_url_hashtag_mention(tweet) for tweet in data.Text])
data = data.assign(url_num = url_num, hash_num = hash_num, mention_num = mention_num)
```

```python
# count number of contractions
contractions = ["'t", "'re", "'s", "'d", "'ll", "'ve", "'m"]
data['contraction_num'] = [sum([tweet.count(cont) for cont in contractions]) for tweet in data.Text]
```

## Text Data Cleaning

This is the most important step of the entire project.  
This cleans the text into a more 'suitable' form as inputs into the NLP models.

* Replace NaNs with 'None'
* Expand Contractions
* Remove Emojis
* Remove URLs
* Remove Punctuations except '!?' as they convey intensity and tonality of tweet
* Replace 'amp' with 'and'
* Word Segmentaion - segment words such as 'iwould' into 'i' and 'would'
* Lemmatization - reduces inflected words into their root form; verb part-of-speech tag is used here)
* Ngrams Exploration 
* Remove Stopwords
* WordCloud of most commmon words (Unigrams)

```python
## Expand Contractions
def decontraction(phrase):
    # specific
    phrase = re.sub(r"won\'t", "will not", phrase)
    phrase = re.sub(r"can\'t", "can not", phrase)

    # general
    phrase = re.sub(r"n\'t", " not", phrase)
    phrase = re.sub(r"\'re", " are", phrase)
    phrase = re.sub(r"\'s", " is", phrase)
    phrase = re.sub(r"\'d", " would", phrase)
    phrase = re.sub(r"\'ll", " will", phrase)
    phrase = re.sub(r"\'t", " not", phrase)
    phrase = re.sub(r"\'ve", " have", phrase)
    phrase = re.sub(r"\'m", " am", phrase)
    return phrase
```

```python
data.Text = [decontraction(tweet) for tweet in data.Text]
```

```python
## Remove Emojis
# Reference : https://gist.github.com/slowkow/7a7f61f495e3dbb7e3d767f97bd7304b
def remove_emoji(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
```

```python
data.Text = data.Text.apply(lambda x: remove_emoji(x))
```

```python
## Remove URLs
data.Text = data.Text.apply(lambda x: remove_url(x))
```

```python
## Remove Punctuations except '!?'
def remove_punct(text):
    new_punct = re.sub('\ |\!|\?', '', punctuation)
    table=str.maketrans('','',new_punct)
    return text.translate(table)

data.Text = data.Text.apply(lambda x: remove_punct(x))
```

```python
## Replace amp
def replace_amp(text):
    text = re.sub(r" amp ", " and ", text)
    return text

data.Text = data.Text.apply(lambda x: replace_amp(x))
```

```python
## Lemmatization
lemmatizer = WordNetLemmatizer()
def lemma(text):
    words = word_tokenize(text)
    return ' '.join([lemmatizer.lemmatize(w.lower(), pos='v') for w in words])

data.Text = data.Text.apply(lambda x: lemma(x))
```

### Ngrams

```python
def generate_ngrams(text, n):
    words = word_tokenize(text)
    return [' '.join(ngram) for ngram in list(get_data(ngrams(words, n))) if not all(w in stop for w in ngram)] # exclude if all are stopwords

def get_data(gen):
    try:
        for elem in gen:
            yield elem
    except (RuntimeError, StopIteration):
        return
```

```python
# Bigrams
bigrams_disaster = data[data.ASD=="ASD"].Text.apply(lambda x: generate_ngrams(x, 2))
bigrams_ndisaster = data[data.ASD=="Non-ASD"].Text.apply(lambda x: generate_ngrams(x, 2))

bigrams_d_dict = {}
for bgs in bigrams_disaster:
    for bg in bgs:
        if bg in bigrams_d_dict:
            bigrams_d_dict[bg] += 1
        else:
            bigrams_d_dict[bg] = 1

bigrams_d_df = pd.DataFrame(bigrams_d_dict.items(), columns=['Bigrams','Count'])

bigrams_nd_dict = {}
for bgs in bigrams_ndisaster:
    for bg in bgs:
        if bg in bigrams_nd_dict:
            bigrams_nd_dict[bg] += 1
        else:
            bigrams_nd_dict[bg] = 1            

bigrams_nd_df = pd.DataFrame(bigrams_nd_dict.items(), columns=['Bigrams','Count'])
```

```python
plt.figure(figsize=(15,10))
plt.subplot(1,2,1)
sns.barplot(x='Count',y='Bigrams',data=bigrams_nd_df.sort_values('Count', ascending=False).head(40), color=color[0]).set_title('Most Common Bigrams for Non-ASD')
ax = plt.gca()
ax.set_ylabel('')
plt.subplot(1,2,2)
sns.barplot(x='Count',y='Bigrams',data=bigrams_d_df.sort_values('Count', ascending=False).head(40), color=color[1]).set_title('Most Common Bigrams for ASD')
ax = plt.gca()
ax.set_ylabel('')
plt.tight_layout()
plt.show()
```

```python
## Remove Stopwords
def remove_stopwords(text):
    word_tokens = word_tokenize(text)
    return ' '.join([w.lower() for w in word_tokens if not w.lower() in stop])

#data_tmp = data.copy()
data['text_nostopwords'] = data.Text.apply(lambda x: remove_stopwords(x))
```

### WordCloud of Most Common Words after Cleaning

Removed some words such as 'new', 'like' and 'people' as they are common between both targets

```python
mask = np.array(Image.open('brain.png'))
reverse = mask[...,::-1,:]

def wc_words(target, sub,mask=mask):
    words = [word.lower() for tweet in data[data.ASD == target].text_nostopwords for word in tweet.split()]
    words = list(filter(lambda w: w != 'toddler', words))
    words = list(filter(lambda w: w != 'toddlers', words))
    dict = {}
    for w in words:
        if w in dict:
            dict[w] += 1
        else:
            dict[w] = 1
    # plot using frequencies        
    my_cloud = WordCloud(background_color='white', stopwords=stop, mask=mask, random_state=0).generate_from_frequencies(dict) 
    
    plt.subplot(1,2,sub)
    plt.imshow(my_cloud, interpolation='bilinear') 
    plt.axis("off")

plt.figure(figsize=(15,10))
wc_words("Non-ASD",1)
plt.title('Non-ASD')
wc_words("ASD", 2,reverse)
plt.title('ASD')
plt.show()
```
